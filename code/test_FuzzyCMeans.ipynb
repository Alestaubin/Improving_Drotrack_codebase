```python
print(res.shape)
```

```python
colour_map = ["#440154FF", "#414487FF", "2A788EFF", "#22A884FF", "#7AD151FF", "#FDE725FF"]

def to_rgb(hex : str):
    substr = hex.replace("#", "")

    r = int(substr[0:2], 16)
    g = int(substr[2:4], 16)
    b = int(substr[4:6], 16)

    return [r, g, b]
```

```python
colored_segmented_image = np.zeros_like(image)

for i in range(5):
        colored_segmented_image[res == i] = np.array(to_rgb(colour_map[i]))
```

```python
fig = plt.figure()
plt.axis("off")
plt.imshow(colored_segmented_image[:, :, ::])
plt.show()
```

```python
import matplotlib.patches as mpatches
from segmentation.segmentation_utils import image_bbox
```

```python
regions = image_bbox(cleaned_up, min_area=500000)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.imshow(cleaned_up)

for region in regions:

    minr, minc, maxr, maxc = region.bbox
    rect = mpatches.Rectangle(
            (minc, minr),
            maxc - minc,
            maxr - minr,
            fill=False,
            edgecolor='red',
            linewidth=2,
    )
    ax.add_patch(rect)

ax.set_axis_off()
plt.tight_layout()
plt.show()
```
